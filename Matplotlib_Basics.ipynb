### The two interfaces on Matplotlib

Due to Matplotlibs historical connection to MATLAB, but also Python's object-oriented approach, there are two interfaces

1. MATLAB- style
    
    Notice here there are no objects createdd - no "="
    
    makes simple plots easy, but more difficult ones tough

```python
plt.plot([1, 2, 3, 4], [3, 4, 2 ,1])
```

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

#### Alternatively, there is object- oriented approach

We will create our fig and ax initially

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [3, 4, 2, 1])
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

Axes objects contain two or three axis objects

We'll create both a figure and axes using subplots()

```python
fig, ax = plt.subplots()
```

This allows us to interact with each element individually

here, pass our x and y data to ax to plot 

```python
x, y = [1, 2, 3, 4], [3, 4, 2, 1]
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

# 3. Line Plots and Style

change appearance of lines by adding to ax.plot

```python
fig, ax = plt.subplots()

#ax.plot(x, y,'o')
ax.plot(x, y, 'r')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y,'o')
#ax.plot(x, y)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.show()
```

we can change the x and y limits

```python
fig, ax = plt.subplots()

ax.plot(x, y,'X')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

We can change the line style and points and width

more on that at:

https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py

```python
fig, ax = plt.subplots()

ax.plot(x, y,'--rX')

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y,'--rX', linewidth = 20)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y,'--ro', linewidth = .5, markerfacecolor = 'aqua', markeredgecolor = 'orchid', 
        markeredgewidth = 4, markersize = 20)

ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.show()
```

### Additional sets of data

```python
fig, ax = plt.subplots()

x = [1, 2, 3, 4]
y = [3, 4, 2, 1]
a = [.5, 1.5, 2.5, 3.5]
b = [3.5, 1.5, .5, 2.5]

ax.plot(x, y,'purple')
ax.plot(a, b, c = "brown", ls = '-.')
        
        
ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.xlim(0, 5)
plt.ylim(0, 5)

plt.grid(False)

plt.show()
```

### Tickmarks

```python
fig, ax = plt.subplots()

x = [1, 2, 3, 4]
y = [3, 4, 2, 1]
a = [.5, 1.5, 2.5, 3.5]
b = [3.5, 1.5, .5, 2.5]

ax.plot(x, y,'purple')
ax.plot(a, b, c = "brown", ls = '-.')
        
        
ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.xlim(0, 5)
plt.ylim(0, 5)

plt.xticks([1, 2, 3])
plt.yticks([1, 4])

plt.grid(False)

plt.show()
```

```python
fig, ax = plt.subplots()

x = [1, 2, 3, 4]
y = [3, 4, 2, 1]
a = [.5, 1.5, 2.5, 3.5]
b = [3.5, 1.5, .5, 2.5]

ax.plot(x, y,'purple')
ax.plot(a, b, c = "brown", ls = '-.')
        
        
ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.xlim(0, 5)
plt.ylim(0, 5)

plt.xticks([1, 2, 3],[ 'Cheap', 'Good', 'Expensive'])
plt.yticks([1, 4], ['Sad', 'Happy'])

plt.grid(False)

plt.show()
```

```python
fig, ax = plt.subplots()

x = [1, 2, 3, 4]
y = [3, 4, 2, 1]
a = [.5, 1.5, 2.5, 3.5]
b = [3.5, 1.5, .5, 2.5]

ax.plot(x, y,'purple')
ax.plot(a, b, c = "brown", ls = '-.')
        
        
ax.set_title("My figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")

plt.xlim(0, 5)
plt.ylim(0, 5)

plt.xticks([1, 2, 3])
plt.yticks([1, 4])

plt.grid(False)

ax.legend(['Car 1', 'Car 2'])

plt.show()
```

# 5. Pie Charts


```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas 

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
counts = titanic['pclass'].value_counts()
```

```python
counts
```

```python
fig, ax = plt.subplots()

ax.pie(counts)

plt.show()
```

```python
print(counts.index)
```

```python
counts = counts.sort_index()

counts.index
```

```python
fig, ax = plt.subplots()

ax.pie(counts, labels = ['1st', '2nd', '3rd'])

plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(counts, autopct = '%1.1f%%')
       
ax.legend(labels = ['1st', '2nd', '3rd'], title = "Titanic Class", loc = 'lower right')

plt.show()
```

# 6. Scatterplots 

```python
titanic_notna = titanic[titanic['age'].notna()]

fig, ax = plt.subplots()

ax.scatter('age', 'fare', data = titanic_notna)

plt.show()
```

```python
female = titanic_notna[titanic_notna['sex'] == 'female']
male = titanic_notna[titanic_notna['sex'] == 'male']

fig, ax = plt.subplots()

ax.scatter('age', 'fare', data = titanic_notna)

scatter_f = ax.scatter('age', 'fare', marker = 'o', c = 'pclass',
                      label = 'female', cmap = 'Paired', alpha = .8,
                      data = female)

scatter_m = ax.scatter('age', 'fare', marker = '^', c = 'pclass',
                      label = 'male', cmap = 'Paired', alpha = .8,
                      data = male)

plt.title("Titanic graph")

ax.set(xlabel = 'Age', ylabel = 'Fare')

ax.legend(*scatter_f.legend_elements())


plt.show()
```

```python
plt.style.use('ggplot')

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True, figsize = (10, 10))


scatter_fd = ax1.scatter('age', 'fare', c = 'pclass',
                      label = 'female', cmap = 'Paired', alpha = .8,
                      data = female[female['survived'] == 0])

scatter_fs = ax2.scatter('age', 'fare', c = 'pclass',
                      label = 'female', cmap = 'Paired', alpha = .8,
                      data = female[female['survived'] == 1])

scatter_md = ax3.scatter('age', 'fare', c = 'pclass',
                      label = 'male', cmap = 'Paired', alpha = .8,
                      data = male[male['survived'] == 0])

scatter_ms = ax4.scatter('age', 'fare', c = 'pclass',
                      label = 'male', cmap = 'Paired', alpha = .8,
                      data = male[male['survived'] == 1])


plt.suptitle("Titanic graph", fontsize = 30)

ax1.set(xlabel = 'Age', ylabel = 'Fare', title = 'Female Died')
ax2.set(xlabel = 'Age', ylabel = 'Fare', title = 'Female Survived')
ax3.set(xlabel = 'Age', ylabel = 'Fare', title = 'Male Died')
ax4.set(xlabel = 'Age', ylabel = 'Fare', title = 'Male Survived')




fig.legend(*scatter_f.legend_elements())


plt.show()
```
